1. Ảnh hưởng của LLMs đến giá trị của việc viết bài luận ở bậc phổ thông?



LLMs là một hệ thống AI được huấn luyện để hiểu, sinh và xử lý ngôn ngữ tự nhiên, điều này này ảnh hưởng rất tích cực với con người và đặc biệt hơn nữa là đối với bậc phổ thông. Học sinh có thể vận dụng tích cực vào việc viết bài luận chẳng hạn như: gợi ý dàn ý, cách triển khai giúp cho học sinh có khung để làm bài luận dễ và nhanh hơn. Nhìn cách LLM viết giúp học sinh học thêm từ vựng, cấu trúc câu, cách hành văn. Và điều có lẽ quan trọng nhất làm cho nhiều học sinh dùng LLM trong việc viết luận văn là sự tiện lợi, thay vì chờ giáo viên LLM có thể chỉ ra lỗi ngữ pháp, mạch lạc, logic gần như tức thì. Tuy nhiên không phải LLM lúc nào cũng tích cực với tất cả học sinh. Dễ nhận thấy ở việc nếu học sinh chỉ copy từ LLM thì sẽ mất đi khả năng tư duy phản biện và kỹ năng lập luận, đây là vấn đề cốt lõi khi viết bài luận. Giáo viên sẽ khó biết học sinh có tự viết hay không, từ đó giảm của điểm số thực của từng học sinh. Và bài viết LLM thường trôi chảy nhưng bị thiếu đi phong cách viết của từng cá nhân nếu dựa vào nó. 

2. LLMs viết code máy tính. Điều này có ý nghĩa gì đối với giá trị của việc học lập trình.

LLMs là một công cụ AI có khả năng sinh, giải thích và tối ưu hóa mã lập trình. Điều này mang lại nhiều tác động tích cực đối với con người và đặc biệt quan trọng trong giáo dục tin học. Học sinh, sinh viên hay lập trình viên có thể tận dụng để tăng năng suất như: gợi ý đoạn code mẫu, phát hiện lỗi, hoặc gợi ý cách triển khai một thuật toán. Nhờ đó, quá trình học lập trình trở nên dễ dàng hơn, giúp người học có thể tiếp cận nhanh hơn với các khái niệm mới và thực hành nhiều hơn. LLM viết code còn giúp người học bổ sung kiến thức như: học được cú pháp chuẩn, cách tổ chức chương trình, và các mẫu thiết kế phổ biến. Điều tiện lợi nhất là LLM có thể hỗ trợ gần như tức thì, thay vì chờ giảng viên hoặc bạn bè giải đáp lỗi, từ đó tiết kiệm thời gian học tập và làm việc. Tuy nhiên, không phải lúc nào LLM cũng mang lại giá trị tích cực cho giáo dục tin học. Nếu người học copy paste từ LLM mà không hiểu bản chất, họ sẽ mất đi khả năng tư duy logic, phân tích và giải quyết vấn đề. Thêm vào đó, mã do LLM sinh ra đôi khi không chính xác, thiếu tối ưu hoặc tiềm ẩn lỗi bảo mật. Vì vậy người không có nền tảng lập trình sẽ khó đánh giá chất lượng code. Một vấn đề khác là dễ khiến người học phụ thuộc, làm giảm khả năng sáng tạo và phong cách lập trình riêng của từng cá nhân.

3. Khi nào học sinh nên được phép sử dụng các công cụ sau? Hãy nêu lý do cho quyết định của bạn.

a. Máy tính bỏ túi

b. LLM (để trả lời bài tập về nhà và viết bài luận)

c. LLM để viết hoặc hỗ trợ viết mã


a. Máy tính bỏ túi nên được phép sử dụng khi học sinh đã nắm chắc các kỹ năng tính toán cơ bản (cộng, trừ, nhân, chia). Vì là nếu cho dùng ngay từ đầu, học sinh sẽ phụ thuộc và mất khả năng rèn luyện tư duy số học. Nhưng khi học đến các môn nâng cao hơn như đại số, lượng giác, xác suất, thống kê thì việc dùng máy tính là cần thiết để tập trung vào tư duy giải quyết vấn đề thay vì chỉ làm thủ công.

b. LLMs nên được dùng như một công cụ hỗ trợ chứ không phải thay thế hoàn toàn. Ví dụ như học sinh có thể dùng để gợi ý dàn ý, kiểm tra lỗi ngữ pháp, gợi ý từ vựng hoặc cách triển khai trong bài luận. Tuy nhiên, việc copy nguyên văn từ LLM là không nên vì sẽ làm mất đi kỹ năng tư duy phản biện, lập luận và phong cách viết cá nhân. Giáo viên có thể cho phép dùng LLM trong quá trình học tập ở nhà nhưng vẫn cần những bài kiểm tra trên lớp để đảm bảo học sinh có kỹ năng thực sự.

c. LLMs nên được cho phép trong việc học lập trình với vai trò như một trợ giảng như: gợi ý code mẫu, phát hiện lỗi, hoặc giải thích một đoạn chương trình. Điều này giúp học sinh tiết kiệm thời gian và học nhanh hơn. Tuy nhiên học sinh phải tự viết và hiểu code, sau đó có thể dùng LLM để kiểm tra hoặc tối ưu. Nếu cho phép sử dụng không kiểm soát, học sinh có thể chỉ copy paste mà không hiểu bản chất, dẫn đến mất kỹ năng phân tích và giải quyết vấn đề.
